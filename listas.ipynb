```python
# Definir la clase Nodo que representa cada elemento de la lista
class Nodo:
  # Inicializar el nodo con un valor y una referencia al siguiente nodo
  def __init__(self, valor, siguiente=None):
    self.valor = valor
    self.siguiente = siguiente

  # Devolver el valor del nodo
  def get_valor(self):
    return self.valor

  # Devolver la referencia al siguiente nodo
  def get_siguiente(self):
    return self.siguiente

  # Cambiar el valor del nodo
  def set_valor(self, nuevo_valor):
    self.valor = nuevo_valor

  # Cambiar la referencia al siguiente nodo
  def set_siguiente(self, nuevo_siguiente):
    self.siguiente = nuevo_siguiente

# Definir la clase ListaLigada que representa la lista completa
class ListaLigada:
  # Inicializar la lista con una referencia a la cabeza y el tamaño
  def __init__(self):
    self.cabeza = None
    self.tamaño = 0

  # Devolver el tamaño de la lista
  def get_tamaño(self):
    return self.tamaño

  # Devolver el valor de la cabeza de la lista
  def get_cabeza(self):
    if self.cabeza is None:
      return None
    else:
      return self.cabeza.get_valor()

  # Devolver el valor de la cola de la lista
  def get_cola(self):
    if self.cabeza is None:
      return None
    else:
      # Recorrer la lista hasta llegar al último nodo
      actual = self.cabeza
      while actual.get_siguiente() is not None:
        actual = actual.get_siguiente()
      return actual.get_valor()

  # Devolver el valor de un nodo por su índice
  def get_valor_por_indice(self, indice):
    if indice < 0 or indice >= self.tamaño:
      return None
    else:
      # Recorrer la lista hasta llegar al nodo deseado
      actual = self.cabeza
      for i in range(indice):
        actual = actual.get_siguiente()
      return actual.get_valor()

  # Insertar un valor al principio de la lista
  def insertar_al_principio(self, valor):
    # Crear un nuevo nodo con el valor y la referencia a la cabeza actual
    nuevo_nodo = Nodo(valor, self.cabeza)
    # Cambiar la cabeza de la lista al nuevo nodo
    self.cabeza = nuevo_nodo
    # Incrementar el tamaño de la lista
    self.tamaño += 1

  # Insertar un valor al final de la lista
  def insertar_al_final(self, valor):
    # Crear un nuevo nodo con el valor y sin referencia al siguiente nodo
    nuevo_nodo = Nodo(valor)
    if self.cabeza is None:
      # Si la lista está vacía, el nuevo nodo es la cabeza
      self.cabeza = nuevo_nodo
    else:
      # Si no, recorrer la lista hasta llegar al último nodo y cambiar su referencia al nuevo nodo
      actual = self.cabeza
      while actual.get_siguiente() is not None:
        actual = actual.get_siguiente()
      actual.set_siguiente(nuevo_nodo)
    # Incrementar el tamaño de la lista
    self.tamaño += 1

  # Insertar un valor en una posición determinada de la lista
  def insertar_por_indice(self, valor, indice):
    if indice < 0 or indice > self.tamaño:
      return False
    elif indice == 0:
      # Si el índice es cero, insertar al principio de la lista
      self.insertar_al_principio(valor)
      return True
    elif indice == self.tamaño:
      # Si el índice es igual al tamaño, insertar al final de la lista
      self.insertar_al_final(valor)
      return True
    else:
      # Si no, recorrer la lista hasta llegar al nodo anterior al índice deseado y cambiar su referencia al nuevo nodo con el valor y la referencia
            # al nodo siguiente al índice deseado
      actual = self.cabeza
      for i in range(indice - 1):
        actual = actual.get_siguiente()
      nuevo_nodo = Nodo(valor, actual.get_siguiente())
      actual.set_siguiente(nuevo_nodo)
      # Incrementar el tamaño de la lista
      self.tamaño += 1
      return True

  # Eliminar un nodo por su índice
  def eliminar_por_indice(self, indice):
    if indice < 0 or indice >= self.tamaño:
      return False
    elif indice == 0:
      # Si el índice es cero, eliminar la cabeza de la lista y cambiarla por el siguiente nodo
      self.cabeza = self.cabeza.get_siguiente()
      # Decrementar el tamaño de la lista
      self.tamaño -= 1
      return True
    else:
      # Si no, recorrer la lista hasta llegar al nodo anterior al índice deseado y cambiar su referencia al nodo siguiente al índice deseado
      actual = self.cabeza
      for i in range(indice - 1):
        actual = actual.get_siguiente()
      actual.set_siguiente(actual.get_siguiente().get_siguiente())
      # Decrementar el tamaño de la lista
      self.tamaño -= 1
      return True

  # Mostrar la lista como una cadena de texto
  def mostrar(self):
    # Crear una lista vacía para almacenar los valores de los nodos
    valores = []
    # Recorrer la lista y agregar los valores a la lista
    actual = self.cabeza
    while actual is not None:
      valores.append(str(actual.get_valor()))
      actual = actual.get_siguiente()
    # Devolver la lista como una cadena separada por comas
    return ", ".join(valores)
```

```python
# Crear una lista ligada vacía
lista = ListaLigada()
# Insertar algunos valores al principio y al final de la lista
lista.insertar_al_principio(10)
lista.insertar_al_final(20)
lista.insertar_al_final(30)
# Mostrar la lista
print(lista.mostrar())
# Salida: 10, 20, 30
# Insertar un valor en la posición 1 de la lista
lista.insertar_por_indice(15, 1)
# Mostrar la lista
print(lista.mostrar())
# Salida: 10, 15, 20, 30
# Eliminar el valor en la posición 2 de la lista
lista.eliminar_por_indice(2)
# Mostrar la lista
print(lista.mostrar())
# Salida: 10, 15, 30
```

```python
# Definir la clase Nodo que representa cada elemento de la lista
class Nodo:
  # Inicializar el nodo con un valor y una referencia al nodo anterior y al nodo siguiente
  def __init__(self, valor, anterior=None, siguiente=None):
    self.valor = valor
    self.anterior = anterior
    self.siguiente = siguiente

  # Devolver el valor del nodo
  def get_valor(self):
    return self.valor

  # Devolver la referencia al nodo anterior
  def get_anterior(self):
    return self.anterior

  # Devolver la referencia al nodo siguiente
  def get_siguiente(self):
    return self.siguiente

  # Cambiar el valor del nodo
  def set_valor(self, nuevo_valor):
    self.valor = nuevo_valor

  # Cambiar la referencia al nodo anterior
  def set_anterior(self, nuevo_anterior):
    self.anterior = nuevo_anterior

  # Cambiar la referencia al nodo siguiente
  def set_siguiente(self, nuevo_siguiente):
    self.siguiente = nuevo_siguiente

# Definir la clase ListaLigada que representa la lista completa
class ListaLigada:
  # Inicializar la lista con una referencia a la cabeza y a la cola y el tamaño
  def __init__(self):
    self.cabeza = None
    self.cola = None
    self.tamaño = 0

  # Devolver el tamaño de la lista
  def get_tamaño(self):
    return self.tamaño

  # Devolver el valor de la cabeza de la lista
  def get_cabeza(self):
    if self.cabeza is None:
      return None
    else:
      return self.cabeza.get_valor()

  # Devolver el valor de la cola de la lista
  def get_cola(self):
    if self.cola is None:
      return None
    else:
      return self.cola.get_valor()

  # Devolver el valor de un nodo por su índice
  def get_valor_por_indice(self, indice):
    if indice < 0 or indice >= self.tamaño:
      return None
    else:
      # Recorrer la lista hasta llegar al nodo deseado
      actual = self.cabeza
      for i in range(indice):
        actual = actual.get_siguiente()
      return actual.get_valor()

  # Insertar un valor al principio de la lista
  def insertar_al_principio(self, valor):
    # Crear un nuevo nodo con el valor dado y la referencia a la cabeza de la lista
    nuevo_nodo = Nodo(valor, None, self.cabeza)
    # Si la lista está vacía, asignar el nuevo nodo como la cabeza y la cola de la lista
    if self.cabeza is None:
      self.cabeza = nuevo_nodo
      self.cola = nuevo_nodo
    else:
      # Si no, cambiar la referencia del nodo anterior de la cabeza de la lista al nuevo nodo y asignar el nuevo nodo como la cabeza de la lista
      self.cabeza.set_anterior(nuevo_nodo)
      self.cabeza = nuevo_nodo
    # Incrementar el tamaño de la lista
    self.tamaño += 1

  # Insertar un valor al final de la lista
  def insertar_al_final(self, valor):
    # Crear un nuevo nodo con el valor dado y la referencia a la cola de la lista
    nuevo_nodo = Nodo(valor, self.cola, None)
    # Si la lista está vacía, asignar el nuevo nodo como la cabeza y la cola de la lista
    if self.cola is None:
      self.cabeza = nuevo_nodo
      self.cola = nuevo_nodo
    else:
      # Si no, cambiar la referencia del nodo siguiente de la cola de la lista al nuevo nodo y asignar el nuevo nodo como la cola de la lista
      self.cola.set_siguiente(nuevo_nodo)
      self.cola = nuevo_nodo
    # Incrementar el tamaño de la lista
    self.tamaño += 1

  # Insertar un valor por su índice en la lista
  def insertar_por_indice(self, valor, indice):
    if indice < 0 or indice > self.tamaño:
      return False
    elif indice == 0:
      # Si el índice es cero, insertar al principio de la lista
      self.insertar_al_principio(valor)
      return True
    elif indice == self.tamaño:
      # Si el índice es igual al tamaño de la lista, insertar al final de la lista
      self.insertar_al_final(valor)
      return True
    else:
      # Si no, recorrer la lista hasta llegar al nodo anterior al índice deseado y crear un nuevo nodo con el valor dado y las referencias al nodo anterior y al nodo siguiente al índice deseado
      actual = self.cabeza
      for i in range(indice - 1):
        actual = actual.get_siguiente()
      nuevo_nodo = Nodo(valor, actual, actual.get_siguiente())
      # Cambiar las referencias de los nodos anterior y siguiente al nuevo nodo
      actual.set_siguiente(nuevo_nodo)
      actual.get_siguiente().set_anterior(nuevo_nodo)
      # Incrementar el tamaño de la lista
      self.tamaño += 1
      return True

  # Eliminar un valor por su índice en la lista
  def eliminar_por_indice(self, indice):
    if indice < 0 or indice >= self.tamaño:
      return False
    elif indice == 0:
      # Si el índice es cero, eliminar la cabeza de la lista y cambiarla por el siguiente nodo
      self.cabeza = self.cabeza.get_siguiente()
      # Si hay un siguiente nodo, cambiar su referencia al nodo anterior a None
      if self.cabeza is not None:
        self.cabeza.set_anterior(None)
      else:
        # Si no hay un siguiente nodo, asignar None a la cola de la lista también
        self.cola = None
      # Decrementar el tamaño de la lista
      self.tamaño -= 1
      return True
    elif indice == self.tamaño - 1:
    # Si el índice es igual al tamaño de la lista menos uno, eliminar la cola de la lista y cambiarla por el anterior nodo
      self.cola = self.cola.get_anterior()
      # Si hay un anterior nodo, cambiar su referencia al nodo siguiente a None
      if self.cola is not None:
        self.cola.set_siguiente(None)
      else:
        # Si no hay un anterior nodo, asignar None a la cabeza de la lista también
        self.cabeza = None
      # Decrementar el tamaño de la lista
      self.tamaño -= 1
      return True
    else:
      # Si no, recorrer la lista hasta llegar al nodo deseado y cambiar las referencias de los nodos anterior y siguiente al nodo anterior y siguiente al nodo deseado respectivamente
      actual = self.cabeza
      for i in range(indice):
        actual = actual.get_siguiente()
      actual.get_anterior().set_siguiente(actual.get_siguiente())
      actual.get_siguiente().set_anterior(actual.get_anterior())
      # Decrementar el tamaño de la lista
      self.tamaño -= 1
      return True

  # Mostrar la lista como una cadena de texto separada por comas
  def mostrar(self):
    # Crear una lista vacía para almacenar los valores de los nodos
    valores = []
    # Recorrer la lista y agregar los valores de los nodos a la lista
    actual = self.cabeza
    while actual is not None:
      valores.append(str(actual.get_valor()))
      actual = actual.get_siguiente()
    # Devolver la lista como una cadena de texto separada por comas
    return ", ".join(valores)
```

```python
# Crear una lista ligada vacía
lista = ListaLigada()
# Insertar algunos valores al principio, al final y por índice
lista.insertar_al_principio(10)
lista.insertar_al_final(20)
lista.insertar_por_indice(15, 1)
lista.insertar_por_indice(5, 0)
lista.insertar_por_indice(25, 4)
# Mostrar la lista
print(lista.mostrar())
# Eliminar algunos valores por índice
lista.eliminar_por_indice(0)
lista.eliminar_por_indice(3)
lista.eliminar_por_indice(1)
# Mostrar la lista
print(lista.mostrar())
```

listas circulares

```python
# Definir la clase Nodo que representa cada elemento de la lista
class Nodo:
    # Inicializar el nodo con un valor y una referencia al siguiente nodo
    def __init__(self, valor, siguiente=None):
        self.valor = valor
        self.siguiente = siguiente

# Definir la clase ListaCircular que representa la lista enlazada
class ListaCircular:
    # Inicializar la lista con un nodo vacío que apunta a sí mismo
    def __init__(self):
        self.vacio = Nodo(None)
        self.vacio.siguiente = self.vacio
        # Usar un atributo para llevar la cuenta del tamaño de la lista
        self.tamaño = 0

    # Definir un método para obtener el valor de un nodo por índice
    def obtener_por_indice(self, indice):
        # Verificar que el índice sea válido
        if indice < 0 or indice >= self.tamaño:
            return None
        # Recorrer la lista hasta llegar al nodo deseado
        nodo = self.vacio.siguiente
        for i in range(indice):
            nodo = nodo.siguiente
        # Devolver el valor del nodo
        return nodo.valor

    # Definir un método para insertar un valor por índice
    def insertar_por_indice(self, valor, indice):
        # Verificar que el índice sea válido
        if indice < 0 or indice > self.tamaño:
            return False
        # Crear un nuevo nodo con el valor dado
        nuevo = Nodo(valor)
        # Recorrer la lista hasta llegar al nodo anterior al deseado
        anterior = self.vacio
        for i in range(indice):
            anterior = anterior.siguiente
        # Enlazar el nuevo nodo con el siguiente y el anterior
        nuevo.siguiente = anterior.siguiente
        anterior.siguiente = nuevo
        # Incrementar el tamaño de la lista
        self.tamaño += 1
        # Devolver True si se insertó correctamente
        return True

    # Definir un método para eliminar un valor por índice
    def eliminar_por_indice(self, indice):
        # Verificar que el índice sea válido
        if indice < 0 or indice >= self.tamaño:
            return False
        # Recorrer la lista hasta llegar al nodo anterior al deseado
        anterior = self.vacio
        for i in range(indice):
            anterior = anterior.siguiente
        # Obtener el nodo a eliminar y su valor
        eliminar = anterior.siguiente
        valor = eliminar.valor
        # Enlazar el nodo anterior con el siguiente y eliminar el nodo
        anterior.siguiente = eliminar.siguiente
        eliminar.siguiente = None
        # Decrementar el tamaño de la lista
        self.tamaño -= 1
        # Devolver el valor eliminado si se eliminó correctamente
        return valor

    # Definir un método para mostrar los valores de la lista como una cadena de texto separada por comas
    def mostrar(self):
        # Crear una lista vacía para almacenar los valores como texto
        valores = []
        # Recorrer la lista y agregar cada valor a la lista de texto
        nodo = self.vacio.siguiente
        for i in range(self.tamaño):
            valores.append(str(nodo.valor))
            nodo = nodo.siguiente
        # Unir los valores con comas y devolver la cadena resultante
        return ", ".join(valores)
```

```python
# Crear una lista circular vacía
lista = ListaCircular()
# Mostrar la lista
print(lista.mostrar())
# Insertar algunos valores por índice
lista.insertar_por_indice(1, 0)
lista.insertar_por_indice(2, 1)
lista.insertar_por_indice(3, 2)
lista.insertar_por_indice(4, 0)
# Mostrar la lista
print(lista.mostrar())
# Obtener algunos valores por índice
print(lista.obtener_por_indice(0))
print(lista.obtener_por_indice(3))
# Eliminar algunos valores por índice
print(lista.eliminar_por_indice(1))
print(lista.eliminar_por_indice(2))
# Mostrar la lista
print(lista.mostrar())
```
